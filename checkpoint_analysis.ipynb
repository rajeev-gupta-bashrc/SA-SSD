```python
import subprocess
import torch

def get_cuda_version():
    try:
        # Run the `nvcc --version` command to get CUDA version
        output = subprocess.check_output(['nvcc', '--version']).decode('utf-8')
        # Extract the version information from the output
        for line in output.split('\n'):
            if 'release' in line:
                version = line.split('release')[-1].strip().split(',')[0]
                return version
    except FileNotFoundError:
        return "CUDA Toolkit not found. Please ensure it is installed and in your PATH."

def check_pytorch_cuda():
    cuda_available = torch.cuda.is_available()
    print(f"CUDA Available: {cuda_available}")

    if cuda_available:
        # Get the current CUDA device
        current_device = torch.cuda.current_device()
        device_name = torch.cuda.get_device_name(current_device)
        print(f"Current CUDA Device: {current_device}")
        print(f"Device Name: {device_name}")

        # Create a tensor and move it to the GPU
        tensor = torch.rand(3, 3).cuda()
        print(f"Tensor on CUDA: {tensor}")

        # Perform a simple operation
        tensor = tensor * 2
        print(f"Tensor after operation: {tensor}")
    else:
        print("CUDA is not available. Please install CUDA and a compatible PyTorch version.")

if __name__ == "__main__":
    # Get and print CUDA Toolkit version
    cuda_version = get_cuda_version()
    print(f"CUDA Toolkit Version: {cuda_version}")

    # Check and verify PyTorch CUDA functionality
    check_pytorch_cuda()
```
